```python
def get_end_year(season):
    # Convert the season to a string (in case it's an integer)
    season_str = str(season)
    
    # Extract the last two digits of the season (e.g., '13' from '1213')
    end_year_last_two_digits = season_str[2:]
    
    # Combine with '20' to form the full year (e.g., '13' -> '2013')
    end_year = '20' + end_year_last_two_digits
    
    # Return the full end year as an integer
    return int(end_year)


seasons = [1213, 1314, 1415, 1516, 1617, 1718]
years = [get_end_year(season) for season in seasons]

print(years)
```

```python
# Apply the function to create the 'Year' column
opera['Season Year'] = opera['season'].apply(get_end_year)

print(opera.head())
```

```python
# Convert the 'Season Year' column to a datetime object (with January 1st of the year)
opera['Season Year'] = pd.to_datetime(opera['Season Year'].astype(str), format='%Y')
```

```python
import pandas as pd

# Create a pandas Series
perf_by_country_season = pd.Series(perf_by_country_season)

# Convert Series to DataFrame
perf_by_country_season = perf_by_country_season.to_frame()

perf_by_country_season.head()
```

```python
perf_by_country_season = perf_by_country_season.reset_index()
perf_by_country_season.head()
```

```python
# Sort by iso and season to ensure the rows are ordered correctly
perf_by_country_season = perf_by_country_season.sort_values(by=['iso', 'season'])

# Calculate the change in performances from the previous season
perf_by_country_season['country_change_from_previous_season'] = perf_by_country_season.groupby('iso')['performances'].diff()

perf_by_country_season.head()
```

```python
# Merge it back to the original dataframe on the 'iso' and 'season' columns
opera = pd.merge(opera, perf_by_country_season[['iso', 'season', 'country_change_from_previous_season']], on=['iso', 'season'], how='left')
opera.head()
```

```python
# Create a pandas Series
perf_by_city_season = pd.Series(perf_by_city_season)

# Convert Series to DataFrame
perf_by_city_season = perf_by_city_season.to_frame()

perf_by_city_season = perf_by_city_season.reset_index()

perf_by_city_season.head()
```

```python
# Sort by iso and season to ensure the rows are ordered correctly
perf_by_city_season = perf_by_city_season.sort_values(by=['iso', 'city', 'season'])

# Calculate the change in performances from the previous season
perf_by_city_season['city_change_from_previous_season'] = perf_by_city_season.groupby(['iso', 'city'])['performances'].diff()

perf_by_city_season.head()
```

```python
# Merge it back to the original dataframe on the 'iso' and 'season' and 'city' columns
opera = pd.merge(opera, perf_by_city_season[['iso', 'city', 'season', 'city_change_from_previous_season']], on=['iso', 'city', 'season'], how='left')
opera.head()
```

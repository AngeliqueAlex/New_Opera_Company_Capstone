```python
#Import pandas, matplotlib.pyplot, and seaborn in the correct lines below
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
opera = pd.read_excel('/Users/angelique/Documents/GitHub/New_Opera_Company_Capstone/opera_new.xlsx')
opera.head()
```

```python
regions = pd.read_csv('/Users/angelique/Documents/GitHub/New_Opera_Company_Capstone/continents2.csv')
regions.head()
```

```python
#Rename the columns to match opera df
regions.rename(columns={'alpha-2': 'iso', 'region': 'continent', }, inplace=True)
```

```python
# Convert the 'iso' column to lowercase
regions['iso'] = regions['iso'].str.lower()
regions.head()
```

```python
# Perform the merge (join) based on 'col1' and 'col2'
opera = pd.merge(opera, regions[['iso', 'continent']], on=['iso'], how='left')
opera.head()
```

```python
# Perform the merge (join) based on 'col1' and 'col2'
opera = pd.merge(opera, regions[['iso', 'sub-region']], on=['iso'], how='left')
opera.head()
```

```python
opera.drop(columns=['Country name EN'], inplace=True)
opera.head()
```

```python
nan_count_per_column = opera.isna().sum()
print(nan_count_per_column)
```

```python
# Filter rows where 'country population' is NaN, select the 'country name' column, and drop duplicates
country_with_nan_continent = opera[opera['continent'].isna()]['Country Name'].drop_duplicates().tolist()

# Show the list of cities without repeats
print(country_with_nan_continent)
```

```python
opera.loc[(opera['Country Name'] == 'United Kingdom'), 'continent'] =  'Europe'
opera.loc[(opera['Country Name'] == 'United Kingdom'), 'sub-region'] =  'Northern Europe'
```

```python
nan_count_per_column = opera.isna().sum()
print(nan_count_per_column)
```

```python
# Save the DataFrame to an Excel file
opera.to_excel('opera_new.xlsx', index=False)
```

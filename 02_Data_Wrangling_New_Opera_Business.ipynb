```python
from fuzzywuzzy import process
```

```python
# Function to find the best fuzzy match for city names from pop in no_population
def fuzzy_match_city(city, choices):
    match = process.extractOne(city, choices)
    return match[0] if match[1] > 80 else None  # Only match if the confidence is greater than 80%

# Apply fuzzy matching to 'city_clean' column of no_population to find corresponding city names in pop
no_population['matched_city'] = no_population['city_clean'].apply(lambda x: fuzzy_match_city(x, popnew['city_clean']))

# Check if 'matched_city' exists and is populated
print(no_population[['city_clean', 'matched_city']])
```
